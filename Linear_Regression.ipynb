```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```
data=pd.read_csv('/content/drive/My Drive/Colab Notebooks/datasets/Admission_Predict.csv')
data=data.drop(['Serial No.'],axis=1)
data
```

```
for i in range(1,7):
  sns.lineplot(x=data.iloc[:,i], y="Chance of Admit ", data=data)
  plt.show()
```

```
for i in range(1,7):
  sns.scatterplot(x=data.iloc[:,i], y="Chance of Admit ", data=data)
  plt.show()
```

```
for i in range(1,8):
  sns.distplot(data.iloc[:,i], kde=True)
  plt.show()
```

```
for i in range(1,7):
  sns.jointplot(x=data.iloc[:,i], y=data['Chance of Admit '], kind="kde")
  plt.show()
```

```

from sklearn.model_selection import train_test_split

X = data.drop(['Chance of Admit '], axis=1)
y = data['Chance of Admit ']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.20, shuffle=False)
```

```
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

model = LinearRegression()
           
```

```
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```
